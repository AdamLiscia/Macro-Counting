```python
# 3231488 rows in data_nutrients
for j in range(3231488):

# grab a value from the output_value column in data_nutrients    
    value = data_nutrients.loc[j, 'Output_value']
    
# now we need to find where to store that value
# 237911 rows in table_to_fill
    for i in range(237911):
        
# if the NDB_No from the value we grabbed matches the NDB_No on the table        
        if table_to_fill.loc[i, 'NDB_No'] == data_nutrients.loc[j, 'NDB_No']:
        
# loop through the columns of the table_to_fill        
            for col_title in table_to_fill:
        
# if the column name matches the nutrient name of the value we grabbed
                if table_to_fill.loc[i, col_title] == data_nutrients.loc[j, 'Nutrient_name']:       

# set that cell equal to the value        
                    table_to_fill.set_value(i, col_title, value)

# if it doesn't match, set it to zero
                else:
                    table_to_fill.set_value(i, col_title, 0)
```
